## Question 2 : Sampling and Stochastic Gradient Descent

# (a) Sample 1 million data points taking values of θ = [θ 0 , θ 1 , θ 2 ] T = [3 1 2] T , and x1 ∼ N (3, 4) and x2 ∼ N (−1, 4) independently, and noise variance in y is given by, σ2 = 2.

```python
import numpy as np
import LinearRegression_SGD as LR
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.metrics as sm
```

```python
reg=LR.LinearRegression(1,2,3)
```

```python
X1=reg.generate_data(mu=3,sd=4,N=1000000)
np.random.seed(30)
sns.distplot(X1, hist=False)
plt.title('Data Distribution of X1')
plt.show()
```

```python
X2=reg.generate_data(mu=-1,sd=4,N=1000000)
sns.distplot(X2, hist=False)
plt.title('Data Distribution of X2')
plt.show()
```

```python
theta=[3,1,2]

Y=1*theta[0]+X1*theta[1]+X2*theta[2] 
Y=Y+reg.generate_data(mu=0,sd=np.sqrt(2),N=1000000)
Y=np.array(Y)
sns.distplot(Y, hist=False)
Y=Y.reshape(-1,1)
plt.show()
```

## $(d)$ 
### In the 3-dimensional parameter space (θj on each axis), plot the movement of θ as the parameters are updated (until convergence) for varying batch sizes. How does the (shape of) movement compare in each case? Does it make intuitive sense? Argue.

```python
def movement(theta,b,f):
    theta1_list=[]
    theta2_list=[]
    for i in range (len(theta)):
        theta1_list.append(theta[i][0])
        theta2_list.append(theta[i][1]) 
        
    theta1_list=np.array(theta1_list,dtype='float64')
    theta2_list=np.array(theta2_list,dtype='float64')
    
    theta0_list=[]
    theta0_list=np.array(b,dtype='float64')

    theta0_list=theta0_list.flatten()
    theta1_list=theta1_list.flatten()
    theta2_list=theta2_list.flatten()
    
    ut0_list=[]
    ut1_list=[]
    ut2_list=[]

    for i in range (len(theta0_list)//f):
        ut0_list.append(theta0_list[i*f])
        ut1_list.append(theta1_list[i*f])
        ut2_list.append(theta2_list[i*f])  
    
    print(len(ut0_list))
    return ut0_list,ut1_list,ut2_list
```

```python
import plotly.graph_objects as go
```
